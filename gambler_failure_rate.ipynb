```python
import numpy as np
import matplotlib.pyplot as plt
import random
position = 100
walk = [position]
steps = 1000
failure = 0
test_count = 1000
walk1 = np.zeros((test_count,steps))
for i in range(test_count):
    walk1[i][0]=100
for j in range(test_count):
    position = 100
    for i in range(steps):
        step = 10 if random.randint(0,1) else -10
        position =position + step
        walk.append(position)
        walk1[j][i]=position
        if position < 0:
            failure = failure + 1
            break
success = test_count - failure
print('success, failure = ',success,',', failure)
plt.figure()
plt.plot(walk[:1000])
plt.plot(walk1[0])

plt.figure()
for i in range(test_count):
    plt.plot(walk1[i])
    
###############################################################
nwalks = 1000
nsteps = 1000
draws = np.random.randint(0,2,size=(nwalks, nsteps))
steps = np.where(draws > 0, 10, -10)
walks = 100 + steps.cumsum(1)

plt.figure()
for i in range(len(walks)):
    plt.plot(walks[i])
```
