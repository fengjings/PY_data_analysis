```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.arange(100)
data1 = np.arange(20)
fig = plt.figure()
data_get = [(20,'first'),
            (40,'second'),
            (80,'third')]
ax = fig.add_subplot(1,1,1)
plt.plot(data,'ko--')
for i,j in data_get:
    ax.annotate(j, xy = (i,data[i]+1),xytext = (i,data[i]+30),
               arrowprops = dict(arrowstyle = 'fancy')
               )

plt.figure()
plt.plot(data1)
```
