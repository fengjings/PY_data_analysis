```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']
path1 = '2020-02-21 data.csv'
data = pd.read_csv(path1, header=0, index_col=0, encoding='utf-8')#读取所有数据
print(data)
print(type(data))
```

```python
display(path1)
display(data)
data1 = pd.DataFrame(data.values.T, index=data.columns, columns=data.index)#转置
display(data1)
```

```python
data1.describe()
```

```python
death_rate=data1['死亡']/data1['累计确诊']
print(death_rate)
print(type(death_rate))
x = death_rate
print(x.describe())
plt.grid()
# plt.ylim(ymax=0.05, ymin=0) #限定Y轴的范围
plt.plot(x,x, 'ro') # ro表示用红色圆点来表示点
plt.show()
death_rate.plot(kind='bar')
plt.show()
```

```python
recover_rate=data1['治愈']/data1['累计确诊']
print(recover_rate)
print(type(recover_rate))
x = recover_rate
print(x.describe())
plt.grid()
# plt.ylim(ymax=0.05, ymin=0) #限定Y轴的范围
plt.plot(x,x, 'ro') # ro表示用红色圆点来表示点
plt.show()
recover_rate.plot(kind='bar')
plt.show()
```
